```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Create a DataFrame
data = {
    'Ename': ['Roshan', 'Amar', 'Ashwini', 'Lohith', 'Mohan', 'Pramod'],
    'Type': ['Regular', 'Adhoc', 'Regular', 'Adhoc', 'Control', 'Regular'],
    'Department': ['CS', 'CS', 'EC', 'EC', 'CS', 'EC'],
    'Experience': [10, 20, 5, 14, 9, 8],
    'Salary': [50000, 15000, 30000, 15000, 10000, 40000]
}
df = pd.DataFrame(data)

# Pivot table for average salary by department
pivot1 = df.pivot_table(values='Salary', index='Department', aggfunc='mean')
print(pivot1)

# Pivot table for sum and mean of salary by type and count of employees
pivot2 = df.pivot_table(values='Salary', index='Type', aggfunc=['sum', 'mean', 'count'])
print(pivot2)

# Visualize with appropriate plots
plt.figure(figsize=(10, 6))
sns.barplot(x='Department', y='Salary', data=df)
plt.title('Salary by Department')
plt.show()

plt.figure(figsize=(10, 6))
sns.barplot(x='Type', y='Salary', data=df)
plt.title('Salary by Type')
plt.show()
```
